#  Separate data into dependent and target variable

```python
X=df3.iloc[:,:-1]
X[:5]
```

```python
Y=df3.iloc[:,-1]
Y[:5]
```

 #  Splitting Dataset

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=12)
```

#  Data Conversion

```python
from sklearn.preprocessing import StandardScaler

ss=StandardScaler()
X_train_ss=ss.fit_transform(X_train)
X_test_ss=ss.transform(X_test)
```

#  Model Creation

#  Linear Regression

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()
lr.fit(X_train_ss,Y_train)
y_pred=lr.predict(X_test_ss)
from sklearn.metrics import r2_score
print("score : ",r2_score(Y_test,y_pred))
```

# Decision Tree 

```python
from sklearn.tree import DecisionTreeRegressor

dtc=DecisionTreeRegressor()
dtc.fit(X_train_ss,Y_train)
Y_pred=dtc.predict(X_test_ss)
from sklearn.metrics import r2_score
print("score : ",r2_score(Y_test,Y_pred))
```

# K Nearest Neighbours

```python
from sklearn.neighbors import KNeighborsRegressor

knn=KNeighborsRegressor()
knn.fit(X_train_ss,Y_train)
Y_pred1=knn.predict(X_test_ss)
from sklearn.metrics import r2_score
print("score : ",r2_score(Y_test,Y_pred1))
```

#  Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

rr=RandomForestRegressor()
rr.fit(X_train_ss,Y_train)
y_pred1=rr.predict(X_test_ss)
from sklearn.metrics import r2_score
print("score : ",r2_score(Y_test,y_pred1))
```

#  Xgboost

```python
import xgboost as xgb

Xgb=xgb.XGBRegressor()
Xgb.fit(X_train_ss,Y_train)
YY_pred=Xgb.predict(X_test_ss)
from sklearn.metrics import r2_score
print("score : ",r2_score(Y_test,YY_pred))
```

# Evaluation

```python
result=pd.DataFrame({
    "Model": ["Linear Regression","Decision Tree","KNN","Random Forest","Xgboost"],
    "Score": [0.421,0.725,0.553,0.879,0.872]})

results = result.sort_values(by="Score",ascending=False)
results = results.set_index("Model")
results
```

Conclusion: Random forest gives best performance so i will use that for further evaluation
